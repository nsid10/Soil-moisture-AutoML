```
import numpy as np
import pandas as pd
```

```
mix = "run_5_with_BOTH"
```

```
df = pd.read_excel(f"/content/gdrive/My Drive/Linux Shared Folder/Soil moisture/models/{mix}/log_h2o_{mix}.xlsx")
```

```
box = df['y_pred']
```

```
box = np.array(box)
```

```
def fixit(box):
    potato = []

    for i in box:
        i = i.strip().split(' ')
        bag = []
        for j in i:
            if '0' in j and '.' in j:
                bag.append(j)
        potato.append(bag)

    final = []
    for i in potato:
        look = []
        for j in i:
            look.append(j[:8])
        final.append(look)
    
    return np.array(final)
```

```
final = fixit(box)
```

```
del df['y_pred']
```

```
y1df = pd.DataFrame(final, columns=[f'y_pred_{i}' for i in range(9)])
```

```
df = pd.concat([df, y1df], axis=1)
```

```
box = df['y_test']
box = np.array(box)
final = fixit(box)
```

```
y2df = pd.DataFrame(final, columns=[f'y_test_{i}' for i in range(9)])
```

```
del df['y_test']
```

```
df = pd.concat([df, y2df], axis=1)
```

```
df.to_excel(f"/content/gdrive/My Drive/Linux Shared Folder/Soil moisture/models/{mix}/log_h2o_{mix}_fixed.xlsx")
```
