```
def normalize(x, mode="minmax", a=None, b=None):
    if mode == "minmax":
        # Min-max [0, 1]
        return (x - x.min(axis=0)) / (x.max(axis=0) - x.min(axis=0))
    if mode == "mean":
        # Mean [-1, 1]
        return (x - x.mean(axis=0)) / (x.max(axis=0) - x.min(axis=0))
    if mode == "custom":
        # Custom range [a, b]
        return a + ((x - x.min(axis=0)) * (b - a)) / (x.max(axis=0) - x.min(axis=0))
```

```
main_path = "/content/gdrive/My Drive/Soil moisture project"

box = os.listdir(f"{main_path}/raw_maps")
```

```
box.sort()
```

```
for n in box:
    im = Image.open(f"{main_path}/raw_maps/{n}")
    map = np.array(im)

    threshold = 1000000

    new = map.copy()
    i = new > threshold
    new[i] = 0.5

    # plt.imshow(new)

    format = 1

    if format == 1:
        im = Image.fromarray(new)
        im.save(f"{main_path}/maps/image_{n[:-4]}.tif")
    elif format == 2:
        new = normalize(new, mode="custom", a=0, b=255)
        im = Image.fromarray(new)
        im = im.convert('RGB')
        im.save(f"{main_path}/maps/image_{n[:-4]}.png")
    else:
        pass
```
