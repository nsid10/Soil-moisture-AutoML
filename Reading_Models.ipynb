```
import h2o
import numpy as np
import os
import pandas as pd

from h2o.automl import H2OAutoML
```

```
h2o.init()
```

```
models = []
```

```
folder_path = '/content/gdrive/My Drive/Linux Shared Folder/Soil moisture/models'

for d in [2, 10, 50]:
    for m in range(12):
        box = os.listdir(f"/content/gdrive/My Drive/Linux Shared Folder/Soil moisture/models/data_{d}_{m}")

        models.append(h2o.load_model(f"/content/gdrive/My Drive/Linux Shared Folder/Soil moisture/models/data_{d}_{m}/{box[0]}"))
```

```
count = 0
for d in [2, 10, 50]:
    for m in range(12):
        print(f"Depth & Feature: {d} cm, mode = {m}, feature = {m + 1}")
        print(models[count])
        count+=1
        print("\n\n\n\n----------------------------------------------------------------------------------------------\n\n\n\n")
```
